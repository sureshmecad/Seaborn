```python
import numpy as np
import seaborn as sns
```

### Histogram

####  1. Generate normally distributed data

```python
data = np.random.randn(1000)
```

```python
sns.distplot(data, kde=True, rug=True)    # rug : Whether to draw a rugplot on the support axis
```

```python
sns.distplot(data, kde=True, rug=False)
```

#### 2. Using dark background with no grids

```python
sns.set_style('dark')
sns.distplot(data, kde=True, rug=True)
```

#### 3. Matplotlib commands can still be applied to seaborn plos

```python
import matplotlib.pyplot as plt
sns.distplot(data, kde=True, rug=True)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```
