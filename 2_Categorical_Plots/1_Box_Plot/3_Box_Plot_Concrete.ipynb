### To Predict the Strength of Concrete

```python
# Read Data
import numpy as np     # Linear Algebra (calculate the mean and standard deviation)
import pandas as pd    # manipulate data, data processing, load csv file I/O (e.g. pd.read_csv)

# Visualization
import seaborn as sns                  # Visualization using seaborn
import matplotlib.pyplot as plt        # Visualization using matplotlib
%matplotlib inline
```

```python
# Import first 5 rows
df = pd.read_csv("F:/03. Suresh/1. Material/05. Data Science/14. Jupyter/08. Data Sets/concrete.csv")
df.head()
```

```python
numerical_features = df.select_dtypes(exclude='object')
numerical_features
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(data=df.strength, palette='winter')
```

```python
plt.subplots(figsize=(12, 6))
ax = sns.boxplot(data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45);
```

```python
plt.figure(figsize=(12, 6))
ax = sns.boxplot(data=df)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(15,10))
plt.boxplot([df['cement'],df['water'],df['slag'],df['age'],df['strength'],df['ash'],df['superplastic'],
                                                 df['coarseagg'],df['fineagg']])
plt.show()
```

```python
# boxplot on numerical features to find outliers
plt.figure(figsize=(20,60), facecolor='white')
plotnumber =1
for numerical_feature in numerical_features:
    ax = plt.subplot(12,3,plotnumber)
    sns.boxplot(df[numerical_feature])
    plt.xlabel(numerical_feature)
    plotnumber+=1
plt.show()
```
