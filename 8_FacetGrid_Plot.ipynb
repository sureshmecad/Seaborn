<h2 style="color:blue" align="left"> FacetGrid Plot </h2>

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

<h2 style="color:blue" align="left"> 1. dataset : "tips"  </h2>

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['day'].value_counts()
```

```python
tips['time'].value_counts()
```

```python
tips['smoker'].value_counts()
```

```python
tips['sex'].value_counts()
```

-------------------------------------------------------------------------------------------------------------------------------

```python
sns.FacetGrid(tips, col='day')
```

```python
sns.FacetGrid(tips, col='sex')
```

```python
x = sns.FacetGrid(tips, row='smoker')
x.map(sns.distplot, 'total_bill')
```

```python
sns.FacetGrid(tips, row='smoker', col='time')
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time')
x = x.map(plt.hist, 'total_bill')
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time')
x = x.map(plt.hist, 'total_bill', color='green')
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time')
x = x.map(plt.hist, 'total_bill', color='blue', bins=15)
```

```python
x = sns.FacetGrid(tips, col='time', row='smoker')
x = x.map(plt.scatter, 'total_bill', 'tip')
```

```python
x = sns.FacetGrid(tips, col='time', row='smoker')
x = x.map(sns.regplot, 'total_bill', 'tip')
```

```python
x = sns.FacetGrid(tips, col='time', row='smoker')
x = x.map(plt.scatter, 'total_bill', 'tip')
```

```python
x = sns.FacetGrid(tips, col='time', hue='smoker')
x = x.map(plt.scatter, 'total_bill', 'tip')
```

```python
x = sns.FacetGrid(tips, col='time', hue='smoker')
x = (x.map(plt.scatter, 'total_bill', 'tip').add_legend())
```

```python
x = sns.FacetGrid(tips, col='day')
x = x.map(sns.boxplot, 'total_bill', 'time')
```

```python
x = sns.FacetGrid(tips, col='day')
x = x.map(sns.boxplot, 'time', 'total_bill')
```

```python
x = sns.FacetGrid(tips, col='day', size=7)
x = x.map(sns.boxplot, 'time', 'total_bill')
```

```python
x = sns.FacetGrid(tips, col='day', size=6, aspect=.4)
x = x.map(sns.boxplot, 'time', 'total_bill')
```

```python
x = sns.FacetGrid(tips, col='day', col_order=['Sat','Thur','Sun','Fri'], size=6, aspect=.4)
x = x.map(sns.boxplot, 'time', 'total_bill', color='red')
```

```python
x = sns.FacetGrid(tips, col='time', hue='smoker', palette='coolwarm')
x = (x.map(plt.scatter, 'total_bill', 'tip').add_legend())
```

```python
x = sns.FacetGrid(tips, col='time', hue='smoker', palette='husl')
x = (x.map(plt.scatter, 'total_bill', 'tip').add_legend())
```

```python
x = sns.FacetGrid(tips, col='time', hue='smoker', palette='winter_r', size=4, aspect=0.7)
x = (x.map(plt.scatter, 'total_bill', 'tip').add_legend())
```
