<h2 style="color:blue" align="left"> Cluster map </h2>

```python
import seaborn as sns
```

### 1. tips

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.clustermap(df.corr(), annot=True, linewidth=2)
```

-------------------------------------------------------------------------------------------------------------------------------

### 2. flights

```python
df1 = sns.load_dataset("flights")
df1.head()
```

```python
df1 = df1.pivot('month','year','passengers')
```

```python
sns.heatmap(df1)
```

```python
sns.clustermap(df1)
```

```python
sns.clustermap(df1.corr(), annot=True, linewidth=2)
```

```python
sns.clustermap(df1.corr(), col_cluster=False)
```

```python
sns.clustermap(df1.corr(), row_cluster=False)
```

```python
sns.clustermap(df1.corr(), cmap='coolwarm', linewidth=2)
```

```python
sns.clustermap(df1.corr(), cmap='Blues_r', linewidths=1)
```

```python
sns.clustermap(df1.corr(), cmap='Blues_r', linewidths=1, figsize=(10,9))
```

```python
# Standardise across row / column
# standard_scale=1 for column
# standard_scale=0 for row

sns.clustermap(df1.corr(), standard_scale=1, linewidths=1)
```

```python
sns.clustermap(df1.corr(), standard_scale=0, linewidths=1)
```

```python
# Normalising dataset by z_score = 0 (rows) / 1 (columns)

sns.clustermap(df1.corr(), z_score=1, linewidths=1)
```

```python
sns.clustermap(df1.corr(), z_score=0, linewidths=1)
```
