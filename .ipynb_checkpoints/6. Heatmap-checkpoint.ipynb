<h2 style="color:blue" align="left"> Heatmap </h2>

```python
import numpy as np
import seaborn as sns
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

<h3 style="color:green" align="left"> 1. flights </h3>

```python
df = sns.load_dataset("flights")
df.head()
```

```python
df = df.pivot('month','year','passengers')
df1 = sns.heatmap(df)
```

```python
df1 = sns.heatmap(df)
plt.xlabel('Year', size=15)
plt.ylabel('Month', size=15)
plt.title('flights: Year Vs Month', size=15)
```

```python
df1 = sns.heatmap(df, annot=True)
```

```python
# fmt : string, optional
#    String formatting code to use when adding annotations.

df1 = sns.heatmap(df, annot=True, fmt='d')
```

```python
df1 = sns.heatmap(df, annot=True, fmt='d', cmap='coolwarm')
```

```python
df1 = sns.heatmap(df, annot=True, fmt='d',cmap="YlGnBu")
```

```python
df1 = sns.heatmap(df, annot=True, fmt='d',cmap="BuPu")
```

```python
df1 = sns.heatmap(df, annot=True, fmt='d',cmap="RdBu")
```

```python
df1 = sns.heatmap(df, annot=True, fmt='d',cmap="RdPu")
```

```python
df1 = sns.heatmap(df, annot=True, fmt='d',cmap="summer")
```

```python
df1 = sns.heatmap(df, annot=True, fmt='d',cmap="winter_r")
```

```python
sns.heatmap(df, center=df.loc["June",1954], annot=True, fmt='d')
```

```python
sns.heatmap(df, center=df.loc['March',1959], annot=True, fmt='d')
```

```python
df1 = sns.heatmap(df, linewidth=2)
```

```python
df1 = sns.heatmap(df, cbar=False)
```

```python
df1 = sns.heatmap(df, linewidths=0.9, annot=True, fmt='d', vmin=100, vmax=650)
```
