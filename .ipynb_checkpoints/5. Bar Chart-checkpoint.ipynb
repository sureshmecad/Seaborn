<h2 style="color:blue" align="left"> Bar Plot (catplot) </h2>

```python
import seaborn as sns
```

<h2 style="color:green" align="left"> 1. iris </h2>

```python
df = sns.load_dataset("iris")
df.head()
```

```python
df.shape
```

```python
df1 = df.head(50)       # Dataset have total rows of 150 & graph comes as very clumpsy. So limited to first 50 rows
sns.catplot(x='sepal_length', y='sepal_width', data=df1, kind='bar')
```

```python
#df1 = df.head(150)       
sns.catplot(x='sepal_length', y='sepal_width', hue='species', data=df1, kind='bar')
```

```python
df1 = df.head(50)
sns.catplot(x='sepal_length', y='sepal_width', hue='species', row='petal_length', data=df1, kind='bar')
```

```python
# Horizontal chart not available in Bar plot. So inverse variable y='', x=''

sns.catplot(y='sepal_length', x='sepal_width', data=df, kind='bar')
```

<h2 style="color:green" align="left"> 2. Superstore Sales </h2>

```python
import pandas as pd
```

```python
sales = pd.read_excel("C:/Users/deepusuresh/Documents/Data Science/01. Python/5. Seaborn/Sample - Superstore Sales.xls")
sales.head()
```

```python
#sales = sales.head(50)       # Dataset have total rows of 150 & graph comes as very clumpsy. So limited to first 50 rows
sns.catplot(x='Product Category', y='Sales', data=sales, kind='bar')
```

```python
sns.catplot(x='Product Category', y='Sales', hue='Ship Mode', data=sales, kind='bar')
```

```python
sns.catplot(x='Product Category', y='Sales', hue='Ship Mode', row='Customer Segment', data=sales, kind='bar')
```

```python
sns.catplot(x='Product Category', y='Sales', hue='Ship Mode', col='Customer Segment', data=sales, kind='bar')
```

```python
sns.catplot(y='Product Category', x='Sales', data=sales, kind='bar')
```
