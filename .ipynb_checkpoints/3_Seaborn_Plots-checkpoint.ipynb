### 1. Visualizing Statistical Relationships

```python
import seaborn as sns
```

```python
df = sns.load_dataset("flights")
df.head()
```

### relplot  --->  Relational Plot

```python
sns.relplot(x='passengers', y='month', data=df)   # Figure-level interface for drawing relational plots onto a FacetGrid.
```

```python
sns.relplot(x='passengers', y='month', hue="year", data=df)
```

```python
df1 = sns.load_dataset("tips")
df1.head()
```

```python
df1.shape
```

```python
sns.distplot(df1.total_bill)         # default bins size & kde plot
```

```python
sns.distplot(df1.total_bill, kde=False, bins=30)
```

```python
sns.distplot(df1.total_bill, kde=True, bins=30)
```

### Pair Plot

```python
sns.pairplot(df1, hue='sex', palette='husl')
```

```python
sns.relplot(x='time', y='tip', kind="line", data=df1)
```

### Strip Plot

```python
sns.stripplot(x='day', y='tip', data=df1)
```

```python
sns.stripplot(x='day', y='tip', data=df1, hue="sex")
```

```python
sns.violinplot(x='day', y='tip', data=df1)
```

### Swarmplot  (strip plot + Violin plot)

```python
sns.swarmplot(x='day', y='tip', data=df1)
```

```python
sns.violinplot(x='day', y='tip', data=df1)
sns.swarmplot(x='day', y='tip', data=df1)
```

### Heatmap

```python
df1.corr()
```

```python
sns.heatmap(df1.corr())
```

```python
sns.heatmap(df1.corr(), annot=True)
```

```python
sns.heatmap(df1.corr(), annot=True, linewidths=3)
```

```python
sns.heatmap(df1.corr(), annot=True, cmap='coolwarm')
```

### cluster map

```python
sns.clustermap(df1.corr(), annot=True, linewidth=2)
```

### Pair Grid Plot

```python
sns.PairGrid(df1)
```

```python
my_var = sns.PairGrid(df1)
my_var.map_diag(sns.distplot)
my_var.map_upper(sns.kdeplot)
my_var.map_lower(sns.boxenplot)
```

### Facet Grid Plot

```python
my_var1 = sns.FacetGrid(df1, col='day')
```

```python
my_var1 = sns.FacetGrid(df1, col='sex')
```

```python
g = sns.FacetGrid(df1, row='smoker')
g.map(sns.distplot, 'total_bill')
```

### Regression Plot

```python
sns.lmplot(x='total_bill', y='tip', data=df1, hue='sex')
```

### 2. Plotting with Categorical data

```python
# catplot : categorical plots
# Categorical scatterplots     (with ``kind="strip"``; the default)

sns.catplot(x='day', y='total_bill', data=df1)    
```

```python
sns.catplot(x='day', y='total_bill', data=df1, kind="point")   
```

```python
sns.catplot(x='day', y='total_bill', data=df1, row="tip", kind="point")  
```

```python
sns.catplot(x='day', y='total_bill', data=df1, col="tip", kind="point")  
```

```python
sns.catplot(x='day', y='total_bill', data=df1, kind="violin")
```

```python
sns.catplot(x='day', y='total_bill', data=df1, kind="boxen")
```

```python
df2 = sns.load_dataset("iris")
df2.head()
```

### Multiplot Grids
#### Use either FacetGrid (or) Pairplot Function

```python
b=sns.FacetGrid(df2, col="species")
b.map(plt.hist, "sepal_length")
```

```python
a = sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

### Plot Aestetics

```python
sns.set(style="dark")
a = sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style="darkgrid")
a = sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style="white", color_codes=True)
a = sns.load_dataset("tips")
sns.boxplot(x='day', y='total_bill', data=a)
```

```python
sns.set(style="white", color_codes=True)
a = sns.load_dataset("tips")
sns.boxplot(x='day', y='total_bill', data=a)
sns.despine(offset=10, trim=True)
```
