### Swarmplot (strip plot + Violin plot)

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x='day', y='tip', data=df)
```

```python
plt.figure(figsize=(10,6))
ax = sns.swarmplot(x="day", y="total_bill", data=df)
```

```python
plt.figure(figsize=(10,6))
ax = sns.swarmplot(x="total_bill", y="day", data=df)
```

```python
sns.set(style="whitegrid")
ax = sns.swarmplot(x = df["total_bill"])
```

```python
plt.figure(figsize=(10,6))
ax = sns.swarmplot(x="day", y="total_bill", hue="sex", data=df)
```

```python
plt.figure(figsize=(10,6))
ax = sns.swarmplot(x="day", y="total_bill", hue="smoker", data=df, palette="Set2", dodge=True)
```

```python
plt.figure(figsize=(10,6))
ax = sns.swarmplot(x="day", y="total_bill", data=df, color="white", edgecolor="gray")
```

```python
df.time.value_counts()
```

```python
plt.figure(figsize=(10,6))
ax = sns.swarmplot(x="time", y="tip", data=df, size=6)
```

```python
plt.figure(figsize=(10,6))
ax = sns.swarmplot(x="time", y="tip", data=df, order=["Dinner", "Lunch"])
```

```python
sns.set(style="whitegrid")
ax = sns.swarmplot(y = df["total_bill"])
```

```python
import pandas as pd
```

```python
data = pd.read_csv("concrete.csv")
data.head()
```

```python
plt.figure(figsize=(10,9))
sns.swarmplot(y=data["cement"], color="orange")
```

```python
# Swarm plot for feature columns
fig,ax = plt.subplots(nrows=4, ncols=2, figsize=(10,18))
plt.suptitle('Violin Plots', fontsize=20, fontweight = 'bold')
sns.swarmplot(y="cement", data=df, ax=ax[0,0], palette='Set3')
sns.swarmplot(y="slag", data=df, ax=ax[0,1], palette='Set3')
sns.swarmplot (y='ash', data=df, ax=ax[1,0], palette='Set3')
sns.swarmplot(y='water', data=df, ax=ax[1,1], palette='Set3')
sns.swarmplot(y="superplastic", data=df, ax=ax[2,0], palette='Set3')
sns.swarmplot(y="coarseagg", data=df, ax=ax[2,1], palette='Set3')
sns.swarmplot (y ='fineagg', data=df, ax=ax[3,0], palette='Set3')
sns.swarmplot(y='age', data=df, ax=ax[3,1], palette='Set3')
plt.show()
```

```python
import itertools
```

```python
cols = ['cement', 'slag', 'ash', 'water', 'superplastic', 'coarseagg', 'fineagg', 'age', 'strength'] 

length = len(cols)

plt.figure(figsize=(12,25))

for i,j in itertools.zip_longest(cols,range(length)):
    
    plt.subplot(3,3,j+1)
    ax = sns.swarmplot( y = data[i],color="orange")
    ax.set_facecolor("k")
    ax.set_ylabel("")
    ax.set_title(i,color="navy")
    plt.subplots_adjust(wspace = .3)
```
